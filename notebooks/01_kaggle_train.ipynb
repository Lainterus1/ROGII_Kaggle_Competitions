```python
# A2a baseline: geometry + GR + DWT + residual target (20 features)
# Auto-detect competition data by finding train/test subdirectories
import os
data_dir = None
for root, dirs, files in os.walk('/kaggle/input'):
    if 'train' in dirs and 'test' in dirs:
        data_dir = root
        break
if data_dir is None:
    raise FileNotFoundError('Competition data (train/test) not found under /kaggle/input')
print('Data dir:', data_dir)

!python scripts/run_train.py \
  --data-dir {data_dir} \
  --n-splits 5 \
  --seed 42 \
  --include-geometry \
  --include-gr \
  --include-gr-dwt \
  --residual-target \
  --output-model /kaggle/working/a2a_lgbm.pkl
```
