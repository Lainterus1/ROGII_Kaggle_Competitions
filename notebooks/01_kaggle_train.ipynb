```python
from pathlib import Path
import os

INPUT_ROOT = Path('/kaggle/input')
COMPETITION_SLUG = 'rogii-wellbore-geology-prediction'

def _path_score(path):
    text = str(path).replace('\\', '/').lower()
    return (0 if COMPETITION_SLUG in text else 1, len(path.parts), text)

candidates = []
for current, dirs, files in os.walk(INPUT_ROOT):
    path = Path(current)
    if 'sample_submission.csv' in files and 'train' in dirs:
        candidates.append(path)
if not candidates:
    raise FileNotFoundError(f'Competition train data not found under {INPUT_ROOT}')
DATA_DIR = sorted(candidates, key=_path_score)[0]
print('Data dir:', DATA_DIR)
```

```python
!python scripts/run_train.py \
  --data-dir {DATA_DIR} \
  --n-splits 5 \
  --seed 42 \
  --include-geometry \
  --include-gr \
  --residual-target \
  --output-model /kaggle/working/baseline_lgbm.pkl
```
