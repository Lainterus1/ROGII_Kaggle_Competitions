```python
import os
# Auto-detect competition data directory
DATA_DIR = None
for root, dirs, _ in os.walk("/kaggle/input"):
    if "train" in dirs and "sample_submission.csv" in os.listdir(root):
        DATA_DIR = root; break
if DATA_DIR is None:
    for root, dirs, _ in os.walk("/kaggle/input"):
        if "train" in dirs:
            DATA_DIR = root; break
if DATA_DIR is None:
    raise FileNotFoundError("Competition data not found")
print("Data dir:", DATA_DIR)
```

```python
# Train A2a: 20 features (geometry + GR + DWT), residual target
!python scripts/run_train.py \
  --data-dir {DATA_DIR} \
  --n-splits 5 \
  --seed 42 \
  --include-geometry \
  --include-gr \
  --include-gr-dwt \
  --residual-target \
  --output-model /kaggle/working/a2a_lgbm.pkl
```

```python
# Predict → submission.csv
!python scripts/run_predict.py \
  --data-dir {DATA_DIR} \
  --model /kaggle/working/a2a_lgbm.pkl \
  --output /kaggle/working/submission.csv
```
